I have noticed an issue where the tokenizer splits words like *munitione* and *multitudine* as if they ended with the enclitic *-ne*.

The way tokenization works is that the `WordTokenizer` in `cltk.tokenize.latin.word` by default consults a list of stop words. These words do not get split, and the list is found in `cltk.tokenize.latin.params` with the name `latin_exceptions`. 

I want to review all the cases of *-ne* in cicero_text_perseus_tokenized.csv and full_data_text_perseus_tokenized.csv so we can add the appropriate words to the stop list.

```python
from csv import DictReader

from cltk.tokenize.latin.params import latin_exceptions

csv_cic = open("./cicero_text_perseus_tokenized.csv", "r", errors="ignore")
csv_caes = open("./full_data_text_perseus_tokenized.csv", "r", errors="ignore")
```

```python
#for when we need to close them
csv_cic.close()
csv_caes.close()
```

If we retokenize each section again, I want to see whether it treats the dash before *ne* separately.

```python
from cltk.tokenize.latin.word import WordTokenizer as LatinWordTokenizer

tokenizer = LatinWordTokenizer()
tokenizer.tokenize("munitio -ne")
```

So it re-tokenizes the dash. In order to get the context. We need to find every 'ne' in the list resulting from tokenization, the previous two tokens, join them together, and add them to a running list.

```python

csv_cic.seek(0)
csv_caes.seek(0)
cic_read = DictReader(csv_cic, quotechar='"')
caes_read = DictReader(csv_caes, quotechar='"')
```

```python
ne_splits = [] 
for row in cic_read:
    retokenized = tokenizer.tokenize(row['tokens'], enclitics_exceptions=latin_exceptions + addtl_exceptions_cicero)
    # Add one because slicing isn't inclusive in Python
    token_indices = [i for i, x in enumerate(retokenized) if x == "ne"]

    if len(token_indices) > 0:
        for i in token_indices:
            token_index = i + 1
            # how to select: [token_index - 3:token_index]
            ne_splits.append(" ".join(retokenized[(token_index - 3):token_index]))
```

```python
ne_splits = set(ne_splits)
ne_splits
```
